## Numerical Computation

### Overflow and Underflow

> Problem stems from the fact that we need to represent infinite number of real numbers with finite bit patterns

**Underflow** happens when numbers near zero are rounded off to zero.

**Overflow** happens when numbers of large magnitude are approximated as $\infty$ or $-\infty$

Example: in softmax function 
- $\frac{exp(x_i)}{\sum_{j=1}^{n}{exp(x_j)}}$, $exp(x_i)$ will underflow if $x_i=c << 0$ 
- and overflow if $x_j=c >> 0$. 
- Both problems can be overcome by evaluating softmax($z$) with $z=x-max_i x_i$. 
- But still numerator can underflow to 0 and if we take log softmax we will get $-\infty$

### Poor Conditioning

- Conditioning refers to how rapidly a function changes in response to input. 
- High conditioning is computationally inefficient because rounding errors in input can lead to large output fluctuations. 
- **Condition Number** for $f(x)=A^{-1}x\quad|\quad \text{ when}\quad A \in \mathbb{R}^{n \times n}$ has an eigenvalue decomposition, is $max_{i,j}\frac{\lambda_i}{\lambda_j}$ where $\lambda_i$ is the largest and $\lambda_j$ the smallest eigenvalue. 
- Poorly conditioned matrices amplify pre-existing errors when we multiply by the true matrix inverse.

### Gradient-Based Optimisation

Usually optimisation $\implies$ finding $x^* = arg\;min(f(x))$ where $f(x)$ can be called 
- objective function
- cost function
- error function
- loss function

We can find $x^*$ by moving in small steps in opposite direction to $f'(x)$

- $f'(x)$=0 at **critical points**
- $f(x)$ at local minimum when it is lower than all neighbouring points.
- $f(x)$ at local maximum when it is higher than all neighbouring points.
- **critical points** that are neither local maxima nor minima are **saddle points**
- $f(x)$ at global maxima/minima when it is at absolute maxima/minima of the function

when $f$ has multiple inputs i.e. $f(\vec{x}) : R^{n} \rightarrow R$, we need partial derivatives $\frac{\partial}{\partial x_i}f(x)$ and the complete vector of partial derivatives is $\nabla_xf(x)$
> in multidimensional setup as above, **critical point** is where every partial derivative is zero

in direction of unit vector $\vec{u}$, the directional derivative is 
$$\begin{align}
&= \frac{\partial}{\partial \alpha}(f(\vec{x}+\alpha \vec{u})) \\
&= \frac{\partial}{\partial \vec{z}}f(\vec{z})(\vec{u})\quad | \quad \vec{z}=\vec{x}+\alpha \vec{u} \quad \text{and applying chain rule}\\
&=\nabla_x{f(x)}\cdot\vec{u}\\
&=u^T\nabla_x{f(x)}
\end{align}
$$

To have fastest descent, we need to move in direction 
$$
\begin{align}
&= arg\;min_{\vec{u}, \vec{u}^T\vec{u}=1}u^T\nabla_x{f(\vec{x})} \\
&= arg\;min_{\vec{u}, \vec{u}^T\vec{u}=1}{||u||}_2{||\nabla_x{f(\vec{x})}||}_2 cos \theta \\
&= arg\;min_{\vec{u}, \vec{u}^T\vec{u}=1}cos \theta \because {||u||}_2=1,\; \nabla_x{f(\vec{x})} \text{ is independent of }\vec{u} \\
&= \vec{u}\text{ is opposite to }\nabla_x f(\vec{x})
\end{align}
$$
Thus we descend fastest in the direction opposite to the gradient $\nabla_x f(\vec{x})$

Every movement is $x'=x-\epsilon \nabla_x f(\vec{x})$ where $\epsilon$ is a chosen fixed step size called **learning rate** which can be set
- as a small constant
- choose so that directional derivative vanishes
- evaluate multiple $\epsilon$  to get smallest value for objective function (**line search**)

### Jacobian and Hessian
matrix of all partial first derivatives is Jacobian, $J_{i,j}=\frac{\partial}{\partial x_j}{f(x)_i}$ while matrix of all second derivatives is Hessian, $H(f)(i,j)=\frac{\partial}{\partial x_i \partial x_j}f(x)_k$

---

#### Worked Example on Jacobian & Hessian

Let $f(\vec{x})=\left[\begin{matrix}x_1^2x_2+x_1x_2^2 \\ 2x_1x_2+x_2^3\end{matrix}\right]$

Then, 
$$
\begin{align}
J_{1,1}&=\frac{\partial}{\partial x_1}f(x)_1=2x_1x_2+x_2^2 \\
J_{1,2}&=\frac{\partial}{\partial x_2}f(x)_1=x_1^2+2x_1x_2 \\
J_{2,1}&=\frac{\partial}{\partial x_1}f(x)_2=2x_2 \\
J_{2,2}&=\frac{\partial}{\partial x_2}f(x)_2=2x_1+3x_2^2 \\
\end{align}
$$

Hence $J=\left[\begin{matrix} 2x_1x_2+x_2^2 & x_1^2+2x_1x_2 \\ 2x_2 & 2x_1+3x_2^2 \end{matrix}\right]$

matrix of all partial second derivatives is Hessian, $H(f)(x)_{i,j}=\frac{\partial}{\partial x_i \partial x_j}f(x)$

For the same $f(\vec{x})$ as above

$$
\begin{align}
H_{1,1,1}&=\frac{\partial^2}{\partial^2 x_1}f(x)_1&=2x_2 \\
H_{1,2,1}&=\frac{\partial^2}{\partial x_1 \partial x_2}f(x)_1&=2x_1+2x_2 \\
H_{2,1,1}&=\frac{\partial^2}{\partial x_2 \partial x_1}f(x)_1&=2x_1+2x_2 \\
H_{2,2,1}&=\frac{\partial^2}{\partial^2 x_2}f(x)_1&=2x_1 \\
H_{1,1,2}&=\frac{\partial^2}{\partial^2 x_1}f(x)_2&=0 \\
H_{1,2,2}&=\frac{\partial^2}{\partial x_1 \partial x_2}f(x)_2&=2 \\
H_{2,1,2}&=\frac{\partial^2}{\partial x_2 \partial x_1}f(x)_2&=2 \\
H_{2,2,2}&=\frac{\partial^2}{\partial^2 x_2}f(x)_2&=6x_2 \\
\end{align}
$$

Hence $H=\left[ \left[ \begin{matrix} 2x_2 & 2x_1+2x_2 \\ 2x_1+2x_2 & 2x_1 \end{matrix} \right] \left[ \begin{matrix} 0 & 2 \\ 2 & 6x_2 \end{matrix}\right]\right]$

---

> Note from the worked example that the *Hessian is a real symmetric matrix*. This is always true whenever the second derivative is continuous.
