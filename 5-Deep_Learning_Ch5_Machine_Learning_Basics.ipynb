# Machine Learning Basics

## Learning Algorithms

### Task $T$
> Learning is the means of attaining the ability to perform the *task*

Types of Task
- **Classification**: produce a function $f\;:\;\mathbb{R}^n \rightarrow \{1,...,k\}$ i.e. specify which of $k$ categories does an input $\epsilon \;\mathbb{R}^n$ belong to; e.g. object recognition
- **Classification with missing Inputs**: learn a set of functions for each subset of missing inputs. One way is to define probability distribution over all relevant variables and then marginalize out missing variables. e.g. medical diagnosis
- **Regression**: produce a function $f\;:\;\mathbb{R}^n \rightarrow \mathbb{R}$ e.g. expected claim of insured person or future price of securities
- **Transcription**: observe relatively unstructured representation of some kind of data and transcribe the information into discrete textual form. e.g. OCR (Google Street View) and speech recognition
- **Machine Translation**: convert sequence of symbols from one language to another
- **Structured Output**: output is a vector, or any other data structure with multiple values, with important relationships between different elements. e.g. producing parse trees etc. This subsumes both transcription and translation as a superset.
- **Anomaly Detection**: flag unusual items from a set of events or objects. e.g. anomalous transaction pattern following credit card theft
- **Synthesis and Sampling**: generate new examples similar to those in training data. e.g. landscape rendering in video games and speech synthesis i.e. audio production from text. The latter is structure output problem but since we want audio to sound natural, we have to use synthesis and sampling techniques.
- **Imputation of missing values**: predict missing entries $x_i$ from new example $x \in \mathbb{R}^n$
- **Denoising**: produce clean example $x \in \mathbb{R}^n$ from corrupted example $\hat{x} \in \mathbb{R}^n$, or find $p(x|\hat{x})$
- **Density Estimation** or **Probability Mass Function estimation**: learn function $p_{model}: \mathbb{R}^n \rightarrow \mathbb{R}$. It can be used to then do imputation.

### Performance Measure $P$

- **Accuracy** i.e. proportion of examples where model gave correct output
- **Error Rate** i.e. proportion of examples where model gives incorrect output. It is also called expected **0-1 loss** where **0-1 loss** for a single example $=0$ when correctly classified and $=1$ when not.
- **Test Set** i.e. apply model to examples it hasn't seen and measure accuracy/error-rate

In practice, it is non-trivial to define performance measure and depends highly on what we want to measure. 

### Experience $E$

- the **dataset** that the learning algorithm is allowed to experience, with examples being also called **data points**.
- **Unsupervised algorithms** experience only dataset (random vectors $x$) and attempt to explcitly or implicitly learn the probability distribution $p(x)$ 
- **Supervised algorithms** experience dataset $x$ with labels $y$ and try to predict $y$ given $x$ usually by learning $p(y|x)$

Lines between Supervised and Unsupervised learning are blurred. 

Using chain rule of probability, for vector $x \in \mathbb{R}^n$, the joint distribution of $x_i$ can be decomposed as $p(x) = \overset{n}{\underset{i=1}{\prod}} p(x_i|x_1,..., x_{i-1}) $ and thus split an unsupervised learning problem into $n$ supervised learning problems. 

Similarly using conditional rule, $p(y\;|\;x)=\frac{p(x,y)}{\sum_{y'}p(x, y')}$ and thus a supervised learning problem gets converted into unsupervised learning problem.

Datasets can be written as **Design Matrices** with rows for each of $m$ examples and columns for each of $n$ features i.e. $X \in \mathbb{R}^{m \times n}$. 

If number of features are not the same for each example, then we write dataset as a set containing m elements $\{x^{(1)},x^{(2)},...,x^{(m)}\}$

### Example: Linear Regression

Input: $x \in \mathbb{R}^n$, output $y \in \mathbb{R}$

$\hat{y}=\vec{w}^T\vec{x}$ where $w$ is vector of **parameters** or **weights**

**Task**: predict $y$ from $x$ by outputting $\hat{y}=\vec{w}^T\vec{x}$

For performance measure, take a test set of examples $X^{(test)}$ and test set of corresponding labels $y^{(test)}$ and measure mean squared error i.e. 
$$MSE_{test}=\frac{1}{m}\underset{i}{\sum}{({\hat{y}}^{(test)}-y^{(test)})}^2_i = \frac{1}{m}{||{\hat{y}}^{(test)}-y^{(test)}||}^2_2$$
Thus, error increases whenever euclidean distance between predictions and targets increases.

To make a machine learing algorithm for this, we need an algorithm that improves $w$ such that $MSE_{test}$ is minimized, as it is given exposure to training set $(X^{(train)},y^{(train)})$

To minimize $MSE_{test}$ we take its gradient to be zero. i.e.
$$
\begin{align}
\nabla_w MSE_{train}&=0 \\
\implies \nabla_w\frac{1}{m}{||{\hat{y}}^{(train)}-y^{(train)}||}^2_2 &= 0 \\
\implies \nabla_w\frac{1}{m}{||X^{(train)}w-y^{(train)}||}^2_2 &= 0 \\
\implies \nabla_w {(X^{(train)}w-y^{(train)})}^T (X^{(train)}w-y^{(train)}) &=0 \\
\implies \nabla_w (w^T X^{(train)T})X^{(train)}w-w^T X^{(train)T} y^{(train)}-y^{(train)T}X^{(train)}w+y^{(train)T}y^{(train)} &=0 \\
\implies \nabla_w (w^T X^{(train)T})X^{(train)}w-w^T X^{(train)T} y^{(train)}-w^T X^{(train)T} y^{(train)}+y^{(train)T}y^{(train)} &=0 \\
\implies \nabla_w (w^T X^{(train)T})X^{(train)}w-2w^T X^{(train)T} y^{(train)}+y^{(train)T}y^{(train)} &=0 \\
\end{align}
$$
