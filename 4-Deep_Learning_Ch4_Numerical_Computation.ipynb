## Numerical Computation

### Overflow and Underflow

> Problem stems from the fact that we need to represent infinite number of real numbers with finite bit patterns

**Underflow** happens when numbers near zero are rounded off to zero.

**Overflow** happens when numbers of large magnitude are approximated as $\infty$ or $-\infty$

Example: in softmax function 
- $\frac{exp(x_i)}{\sum_{j=1}^{n}{exp(x_j)}}$, $exp(x_i)$ will underflow if $x_i=c << 0$ 
- and overflow if $x_j=c >> 0$. 
- Both problems can be overcome by evaluating softmax($z$) with $z=x-max_i x_i$. 
- But still numerator can underflow to 0 and if we take log softmax we will get $-\infty$

### Poor Conditioning

- Conditioning refers to how rapidly a function changes in response to input. 
- High conditioning is computationally inefficient because rounding errors in input can lead to large output fluctuations. 
- **Condition Number** for $f(x)=A^{-1}x\quad|\quad \text{ when}\quad A \in \mathbb{R}^{n \times n}$ has an eigenvalue decomposition, is $max_{i,j}\frac{\lambda_i}{\lambda_j}$ where $\lambda_i$ is the largest and $\lambda_j$ the smallest eigenvalue. 
- Poorly conditioned matrices amplify pre-existing errors when we multiply by the true matrix inverse.

### Gradient-Based Optimisation

Usually optimisation $\implies$ finding $x^* = arg\;min(f(x))$ where $f(x)$ can be called 
- objective function
- cost function
- error function
- loss function

We can find $x^*$ by moving in small steps in opposite direction to $f'(x)$

- $f'(x)$=0 at **critical points**
- $f(x)$ at local minimum when it is lower than all neighbouring points.
- $f(x)$ at local maximum when it is higher than all neighbouring points.
- **critical points** that are neither local maxima nor minima are **saddle points**
- $f(x)$ at global maxima/minima when it is at absolute maxima/minima of the function

when $f$ has multiple inputs i.e. $f(\vec{x}) : R^{n} \rightarrow R$, we need partial derivatives $\frac{\partial}{\partial x_i}f(x)$ and the complete vector of partial derivatives is $\nabla_xf(x)$
> in multidimensional setup as above, **critical point** is where every partial derivative is zero

in direction of unit vector $\vec{u}$, the directional derivative is 
$$\begin{align}
&= \frac{\partial}{\partial \alpha}(f(\vec{x}+\alpha \vec{u})) \\
&= \frac{\partial}{\partial \vec{z}}f(\vec{z})(\vec{u})\quad | \quad \vec{z}=\vec{x}+\alpha \vec{u} \quad \text{and applying chain rule}\\
&=\nabla_x{f(x)}\cdot\vec{u}\\
&=u^T\nabla_x{f(x)}
\end{align}
$$

To have fastest descent, we need to move in direction 
$$
\begin{align}
&= arg\;min_{\vec{u}, \vec{u}^T\vec{u}=1}u^T\nabla_x{f(\vec{x})} \\
&= arg\;min_{\vec{u}, \vec{u}^T\vec{u}=1}{||u||}_2{||\nabla_x{f(\vec{x})}||}_2 cos \theta \\
&= arg\;min_{\vec{u}, \vec{u}^T\vec{u}=1}cos \theta \because {||u||}_2=1,\; \nabla_x{f(\vec{x})} \text{ is independent of }\vec{u} \\
&= \vec{u}\text{ is opposite to }\nabla_x f(\vec{x})
\end{align}
$$
Thus we descend fastest in the direction opposite to the gradient $\nabla_x f(\vec{x})$

Every movement is $x'=x-\epsilon \nabla_x f(\vec{x})$ where $\epsilon$ is a chosen fixed step size called **learning rate** which can be set
- as a small constant
- choose so that directional derivative vanishes
- evaluate multiple $\epsilon$  to get smallest value for objective function (**line search**)

### Jacobian and Hessian
matrix of all partial first derivatives is Jacobian, $J_{i,j}=\frac{\partial}{\partial x_j}{f(x)_i}$ while matrix of all second derivatives is Hessian, $H(f)(i,j)=\frac{\partial}{\partial x_i \partial x_j}f(x)_k$

---

#### Worked Example on Jacobian & Hessian

Let $f(\vec{x})=\left[\begin{matrix}x_1^2x_2+x_1x_2^2 \\ 2x_1x_2+x_2^3\end{matrix}\right]$

Then, 
$$
\begin{align}
J_{1,1}&=\frac{\partial}{\partial x_1}f(x)_1=2x_1x_2+x_2^2 \\
J_{1,2}&=\frac{\partial}{\partial x_2}f(x)_1=x_1^2+2x_1x_2 \\
J_{2,1}&=\frac{\partial}{\partial x_1}f(x)_2=2x_2 \\
J_{2,2}&=\frac{\partial}{\partial x_2}f(x)_2=2x_1+3x_2^2 \\
\end{align}
$$

Hence $J=\left[\begin{matrix} 2x_1x_2+x_2^2 & x_1^2+2x_1x_2 \\ 2x_2 & 2x_1+3x_2^2 \end{matrix}\right]$

matrix of all partial second derivatives is Hessian, $H(f)(x)_{i,j}=\frac{\partial}{\partial x_i \partial x_j}f(x)$

For the same $f(\vec{x})$ as above

$$
\begin{align}
H_{1,1,1}&=\frac{\partial^2}{\partial^2 x_1}f(x)_1&=2x_2 \\
H_{1,2,1}&=\frac{\partial^2}{\partial x_1 \partial x_2}f(x)_1&=2x_1+2x_2 \\
H_{2,1,1}&=\frac{\partial^2}{\partial x_2 \partial x_1}f(x)_1&=2x_1+2x_2 \\
H_{2,2,1}&=\frac{\partial^2}{\partial^2 x_2}f(x)_1&=2x_1 \\
H_{1,1,2}&=\frac{\partial^2}{\partial^2 x_1}f(x)_2&=0 \\
H_{1,2,2}&=\frac{\partial^2}{\partial x_1 \partial x_2}f(x)_2&=2 \\
H_{2,1,2}&=\frac{\partial^2}{\partial x_2 \partial x_1}f(x)_2&=2 \\
H_{2,2,2}&=\frac{\partial^2}{\partial^2 x_2}f(x)_2&=6x_2 \\
\end{align}
$$

Hence $H=\left[ \left[ \begin{matrix} 2x_2 & 2x_1+2x_2 \\ 2x_1+2x_2 & 2x_1 \end{matrix} \right] \left[ \begin{matrix} 0 & 2 \\ 2 & 6x_2 \end{matrix}\right]\right]$

---

> Note from the worked example that the *Hessian is a real symmetric matrix*. This is always true whenever the second derivative is continuous. 

Now we know the second derivative defines the curvature. In the multi-dimensional case, the curvature in any direction defined by unit vector $\vec{d}$ is given as $\vec{d}^TH\vec{d}$ where $H$ is the Hessian.

When $d$ is an eigenvector of $H$ then we know that $H\vec{d}=\lambda\vec{d} \therefore \vec{d}^TH\vec{d}=\vec{d}^T\lambda\vec{d}=\lambda\vec{d}^T\vec{d}=\lambda \because {||d||}_2=1$

Thus
- the curvature in the direction of the eigenvevtors is the eigenvalue corresponding to that eigenvector. 
- Curvature in any other direction is a weighted average of the eigenvalues with weights between 0 and 1. 
- maximum curvature (or 2nd derivative) is in the direction of the eigenvector with $\lambda_{max}$
- minimum curvature (or 2nd derivative) is in the direction of the eigenvector with $\lambda_{min}$



#### Hessian and Gradient Descent Performance

Taylor Series expansion is given as $f(x)=\sum_{n=0}^\infty{\frac{f^(n)a}{n!}{(x-a)}^n}$

Applying it to our function 

First Term: $f(a) = f(x^{(0)})$

Second Term: 
$$
\begin{align}
(x-a)f'(a) &= \\
&= {(x-x^{(0)})}^T \nabla f(x^{(0)}) \because \text{ product must be scalar \& 1st derivative is gradient} \\
&= {(x-x^{(0)})}^T g(x^{(0)}) \text{ (writing }\nabla f(x) \text{ as } g(x) \text{ )}\\
&= {(x-x^{(0)})}^T g \text{ (dropping }(x^{(0)})\text{ for further simplicity)}
\end{align}
$$

Third Term:
$$
\begin{align}
\frac{1}{2!}{(x-a)}^2f'(a) &= \\
&= \frac{1}{2}{(x-x^{(0)})}^T H(x^{(0)}){(x-x^{(0)})} \because \text{ product must be scalar \& 2nd derivative is H} \\
&= \frac{1}{2}{(x-x^{(0)})}^T H {(x-x^{(0)})} \text{ (dropping }(x^{(0)})\text{ for further simplicity)} \\
\end{align}
$$

Combining all three gives us
$$f(x) \approx f(x^{(0)})+{(x-x^{(0)})}^Tg+\frac{1}{2}{(x-x^{(0)})}^T H {(x-x^{(0)})}$$

Now if we use a learning rate of $\epsilon$, the new point is $x = x^{(0)}-\epsilon g \implies x-x^{(0)} = -\epsilon g$

Substituting this, our equation becomes
$$f(x^{(0)}-\epsilon g) \approx f(x^{(0)})-\epsilon g^Tg+\frac{1}{2}{\epsilon g}^T H {\epsilon g}$$
or
$$f(x^{(0)}-\epsilon g) \approx f(x^{(0)})-\epsilon g^Tg+\frac{1}{2}\epsilon^2 g^T H g$$
These 3 terms are
1. original function
2. expected improvement due to slope of function
3. correction for curvature of function

- when $g^T H g >> 0$, gradient descent can actually go upwards (reverse of what we want)
- when $g^T H g \le 0$, gradient descent step will become larger as we increase $\epsilon$

In practice, Taylor series may not hold as $\epsilon$ increases, thus we choose optimal $\epsilon$ by minimising $f(x^{(0)}-\epsilon g)$ w.r.t. $\epsilon$

We get 
$$
\begin{align}
\frac{\partial}{\partial \epsilon}(x^{(0)}-\epsilon g)&= 0 \\
\implies \frac{\partial}{\partial \epsilon}f(x^{(0)})-\frac{\partial}{\partial \epsilon}\epsilon g^Tg+\frac{\partial}{\partial \epsilon}\frac{1}{2}\epsilon^2 g^T H g &= 0 \\
\implies -g^Tg+\epsilon g^T H g &= 0 \\
\implies \epsilon=\epsilon^*&=\frac{g^Tg}{g^T H g}
\end{align}
$$
When $g$ is aligned with eigenvector corresponding to the max eigenvalue of $H$, 

we get $g^Tg=1$ (orthonormal) and $g^T H g=\lambda_{max} \implies \epsilon=\frac{1}{\lambda_{max}}$

> Thus, to the extent that the function we are optimizing can be approximated by a quadratic function (Taylor series limit), the eigenvalues of the Hessiam determine the scale of the learning rate

#### Second Derivative Test
In the one dimensional case
- $f'(x)=0$ and $f''(x)>0 \implies$ local minimum
- $f'(x)=0$ and $f''(x)<0 \implies$ local maximum
- $f'(x)=0$ and $f''(x)=0 \implies$ test inconclusive, saddle/flat point

In multi-dimensional case
- $\nabla_xf(x)=0$, $H$ is positive definite i.e. all eigenvalues are positve, $\implies$ **local minima**
- $\nabla_xf(x)=0$, $H$ is negative definite i.e. all eigenvalues are negative, $\implies$ **local maxima**
- $\nabla_xf(x)=0$, $H$ has some eigenvalues positve and some negative, $\implies$ **local minima in one cross-section, local maxima in another cross-section**
- $\nabla_xf(x)=0$, $H$ has some eigenvalues as zero and non-zero eigenvalues have same sign, $\implies$ **test is inconclusive**
- when condition of Hessian is poor i.e. $\frac{\lambda_{max}}{\lambda_{min}}$ is very high, **gradient descent performs poorly** and **choosing step size is difficult**.

#### Newton's Method
Uses second order Taylor Series expansion to approximate $f(x)$ near some point $x^{(0)}$

$$f(x)=f(x^{(0)})+(x-x^{(0)})\nabla_xf(x^{(0)})+\frac{1}{2}{(x-x^{(0)})}^T H(f)(x^{(0)}) (x-x^{(0)})$$

Solving this for critical point, we have to get $\nabla_xf(x)=0$
In the RHS:

First Term

It is a function of $x^{(0)}$ which is a constant w.r.t. $x$ and hence $\nabla_xf(x^{(0)})=0$

Second Term

Let $g=\nabla_xf(x^{(0)})$ which is a constant vector

Then $\nabla_x[(x-x^{(0)})g] = g$

Third Term

let $v=(x-x^{(0)})$

Then, we need $\nabla_x\left[\frac{1}{2}v^T H(f)(x^{(0)})v\right] = Hv = H(x-x^{(0)})$

Thus, overall we have for $\nabla_xf(x)=0$ and using shorthand for $H$ and $g$
$$
\begin{align}
g+H(x^*-x^{(0)})&=0 \\
H(x^*-x^{(0)})&=-g \\
H^{-1}H(x^*-x^{(0)})&=-H^{-1}g \\
x^*&=x^{(0)}-H^{-1}g \\
\end{align}
$$
Thus, expanding formally $x^*=x^{(0)}-H(f)(x^{(0)})\nabla_xf(x^{(0)})$

#### Newton's Method Utility

- If $f(x)$ is a quadratic positive definite function, Newton's method takes it to the local minima in one single jump. 
- If $f(x)$ can be approximated as a positive definite quadratic then Newton's method can iteratively take it to the local minima.
- Newton's method takes $f(x)$ to the minima much faster than Gradient Descent
- Newton's method can be harmful near a saddle point since it is attracted to that also while Gradient Descent is not attracted to a saddle point.

- Optimization algorithms that use only the gradient are called **first-order optimization** algorithms. Example is Gradient Descent
- Optimization algorithm that use the Hessian also are called **second-order optimization** algorithms. Example is Newton's method
- Optimization algorithms take **no guarantee** of finding a local minima/maxima, more so in deep learning where complicated functions are used
- Some guarantee can be had if the function is **Lipschitz Continuous** or has Lipschitz Continuous derivatives.
- Lipschitz continuity is a weak constraint and hence widely applicable
- For convex functions, convex optimization algorithms can provide many more guarantees by putting many more restrictions. But they are not widely applicable in deep learning.

> A **Lipschitz Continuous** function is one whose rate of change is bounded by a **Lipschitz Constant** i.e. $\forall x, \forall y |f(x)-f(y)|\le\mathcal{L}{||x-y||}_2$

> A **Convex Function** is one for which the Hessian is positive semi-definite everywhere. They don't have saddle points and all their local minima are also global minima

### Constrained Optimization

Definition: when we want to find max/min of $f(x)$ such that $x$ is always feasible, i.e. in a set $\mathbb{S}$

Problem that arises is that during optimization, such as gradient descent, we may reach a point outside of $\mathbb{S}$. Solutions
1. Project the resulting value back into $\mathbb{S}$. (see Projection below)
2. If using line search, search only over $\epsilon$ such that yield feasible points, or project the points back into $\mathbb{S}$
3. Project the gradient into the tangent space of $\mathbb{S}$ (see Tangent Space below)
4. Design a different unconstrained optimization problem whose solution can be converted into the original constrained optimization problem. 
5. Use Karush-Kuhn-Tucker approach (see below)

> **Projection**: Mapping a point that lies outside the feasible set S back to the nearest point inside S.

> **Tangent space**: The flat plane at a point on the boundary of S containing all directions that keep you on the boundary to first order — the legal directions of movement.

#### Karush-Kuhn-Tucker (KKT) approach

Define set $\mathbb{S}$ in terms of $m$ equalities $g^{(i)}(x)=0$ and $n$ inequalities $h^{(j)}(x)\le0$

Use $m$ $\lambda_is$ and $n$ $\alpha_js$ to construct the **Lagrangian**

$L(x,\lambda,\alpha)=f(x)+\sum_i \lambda_i g^{(i)}(x)+\sum_j \alpha_j h^{(j)}(x)$

We will now show that by doing unconstrained optimization of the generalised Lagrangian, we can achieve the same result as by doing constrained optimization of $f(x)$

Let this unconstrained optimization be $\underset{x}{min}\,\underset{\lambda}{max}\,\underset{\alpha, \alpha \ge 0}{max}\,L(x,\lambda,\alpha)$ i.e. the Lagrangian is maximised for $\lambda$ and $\alpha$ and minimized for $x$, without any constraints.

If constraints on $f(x)$ are satisfied, then we have $L(x,\lambda,\alpha)=f(x)+0+t \mid t\le 0 $. 

Thus, $\underset{\lambda}{max}\,\underset{\alpha, \alpha \ge 0}{max}\,L(x,\lambda,\alpha)=f(x)$ since our only option is to keep $\alpha_j=0$

If any of the constraints on $f(x)$ are *not* satisfied, then we can always choose the corresponding $\lambda$ or $\alpha$ so we take $L(x,\lambda,\alpha)$ to $\infty$ so that 

$\underset{\lambda}{max}\,\underset{\alpha, \alpha \ge 0}{max}\,L(x,\lambda,\alpha)=\infty$

These properties guarantee that no infeasible point can be optimal, and that the optimum within the feasible points is unchanged. 

To do unconstrained maximization (instead of minimizatio) we can construct the lagrangian for $-f(x)$ to get $L(x,\lambda,\alpha)=-f(x)+\sum_i \lambda_i g^{(i)}(x)+\sum_j \alpha_j h^{(j)}(x)$

The sign of the equalities doesn't matter because we can choose any value of $\lambda$
